```python
# http://books.toscrape.com/
# Visit the Books to Scrape site
url = 'http://books.toscrape.com/'
browser.visit(url)
# Parse the HTML
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Scrape the list of book URLs
#list_books = soup.find('div', class_='image_container' )

# Get the Ordered List subtree, which contains the links to the books on the page.
list_books = soup.find('ol')
list_books
```

```python
# Now get the "a" tags which contain the book relative URLS in HREF attributes
links_in_a_tags = list_books.find_all('a')
links_in_a_tags
```

```python
#for link in soup.find_all('a'):

# For each a tag, get the HREF attribute value, which in this case is a relative URL
# NOTE: the URLS returned are doubled-up, as each book list item contains two URLs to the same book (with one URL for a book's image, the other for a book title)

for link in links_in_a_tags:
    print (link.get('href'))
```
